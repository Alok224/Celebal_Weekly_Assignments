```python
import psycopg2
from psycopg2 import sql

DB_NAME = "hospital_db"
DB_USER = "postgres"
DB_PASSWORD = "12345678"
DB_HOST = "localhost"
DB_PORT = "5432"

# connect to default postgres db first to create hospital_db if needed
conn = psycopg2.connect(dbname="postgres", user=DB_USER, password=DB_PASSWORD, host=DB_HOST, port=DB_PORT)
conn.autocommit = True
cur = conn.cursor()

cur.execute("SELECT 1 FROM pg_database WHERE datname = %s", (DB_NAME,))
db_exists = cur.fetchone()

if not db_exists:
    cur.execute(sql.SQL("CREATE DATABASE {}").format(sql.Identifier(DB_NAME)))
    print(f"Database '{DB_NAME}' created.")
else:
    print(f"Database '{DB_NAME}' already exists.")

cur.close()
conn.close()
```

```python
# connect to hospital_db
conn = psycopg2.connect(dbname=DB_NAME, user=DB_USER, password=DB_PASSWORD, host=DB_HOST, port=DB_PORT)
cur = conn.cursor()
print("Connected to hospital_db")
```

```python
create_tables_query = """
DROP TABLE IF EXISTS insurance;
DROP TABLE IF EXISTS medical_records;
DROP TABLE IF EXISTS admissions;
DROP TABLE IF EXISTS patients;

CREATE TABLE patients (
    patient_id SERIAL PRIMARY KEY,
    name VARCHAR(100),
    age INT,
    gender VARCHAR(20),
    blood_type VARCHAR(10)
);

CREATE TABLE admissions (
    admission_id SERIAL PRIMARY KEY,
    patient_id INT REFERENCES patients(patient_id),
    admission_date DATE,
    discharge_date DATE,
    admission_type VARCHAR(50),
    room_number INT,
    hospital VARCHAR(150),
    doctor VARCHAR(100)
);

CREATE TABLE medical_records (
    record_id SERIAL PRIMARY KEY,
    patient_id INT REFERENCES patients(patient_id),
    medical_condition VARCHAR(100),
    medication VARCHAR(100),
    test_results VARCHAR(50)
);

CREATE TABLE insurance (
    insurance_id SERIAL PRIMARY KEY,
    patient_id INT REFERENCES patients(patient_id),
    insurance_provider VARCHAR(100),
    billing_amount NUMERIC(12, 2)
);
"""

cur.execute(create_tables_query)
conn.commit()
print("Tables created successfully.")
```

```python
insert_patient_query = """
INSERT INTO patients (name, age, gender, blood_type)
VALUES (%s, %s, %s, %s)
RETURNING patient_id;
"""

insert_admission_query = """
INSERT INTO admissions (patient_id, admission_date, discharge_date, admission_type, room_number, hospital, doctor)
VALUES (%s, %s, %s, %s, %s, %s, %s);
"""

insert_medical_query = """
INSERT INTO medical_records (patient_id, medical_condition, medication, test_results)
VALUES (%s, %s, %s, %s);
"""

insert_insurance_query = """
INSERT INTO insurance (patient_id, insurance_provider, billing_amount)
VALUES (%s, %s, %s);
"""

for _, row in df.iterrows():
    cur.execute(insert_patient_query, (row["Name"], row["Age"], row["Gender"], row["Blood Type"]))
    patient_id = cur.fetchone()[0]

    cur.execute(insert_admission_query, (
        patient_id, row["Date of Admission"], row["Discharge Date"],
        row["Admission Type"], row["Room Number"], row["Hospital"], row["Doctor"]
    ))

    cur.execute(insert_medical_query, (
        patient_id, row["Medical Condition"], row["Medication"], row["Test Results"]
    ))

    cur.execute(insert_insurance_query, (
        patient_id, row["Insurance Provider"], row["Billing Amount"]
    ))

conn.commit()
print(f"Inserted {len(df)} patient records across all tables.")
```

```python
create_indexes_query = """
CREATE INDEX IF NOT EXISTS idx_patient_name ON patients(name);
CREATE INDEX IF NOT EXISTS idx_medical_condition ON medical_records(medical_condition);
CREATE INDEX IF NOT EXISTS idx_hospital ON admissions(hospital);
CREATE INDEX IF NOT EXISTS idx_doctor ON admissions(doctor);
CREATE INDEX IF NOT EXISTS idx_insurance_provider ON insurance(insurance_provider);
"""

cur.execute(create_indexes_query)
conn.commit()
print("Indexes created successfully.")
```

```python
cur.execute("""
    SELECT table_name FROM information_schema.tables
    WHERE table_schema = 'public';
""")
print("Tables in hospital_db:")
for table in cur.fetchall():
    print("-", table[0])

for table_name in ["patients", "admissions", "medical_records", "insurance"]:
    print(f"\nSchema for {table_name}:")
    cur.execute("""
        SELECT column_name, data_type FROM information_schema.columns
        WHERE table_name = %s;
    """, (table_name,))
    for col in cur.fetchall():
        print("-", col[0], ":", col[1])
```
